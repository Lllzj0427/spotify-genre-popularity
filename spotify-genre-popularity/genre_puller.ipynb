```python
import os

from dotenv import load_dotenv, find_dotenv
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import requests
from time import sleep
from auth import sp
```

```python
# dotenv_path = find_dotenv(usecwd=True)
# print (dotenv_path )


#print ("us 1")


def search_artist(limit, market, artist_name, artist_id):
    for i in range(0, 100, limit):
        track_results = sp.search(q='year:2022', type='track', limit=limit, offset=i, market= market)
    
        for i, t in enumerate(track_results['tracks']['items']):
            artist_name.append(t['artists'][0]['name'])
            artist_id.append(t['artists'][0]['id'])
            
        sleep(0.1)
        
    return  artist_name, artist_id

def search_genre(artist_id, genres_NA):
    
    for i in artist_id:
        
      
        artist_info = sp.artist(i)
        #print ("us 2")

        artist_genre = artist_info["genres"]

        if len(artist_genre) != 0:
            for x in artist_genre:
                if x in genres_NA:
                    genres_NA[x] +=1
                else:
                    genres_NA[x] =0
                    genres_NA[x] +=1
                    
   

    #print ("us 3")               
    return genres_NA

def genre_total(country_genre, total_genre):
    
    for i in country_genre:
        if i not in total_genre:
            total_genre[i] =0
            
        total_genre[i] +=1
            




    return total_genre





artist_name = []

artist_id = []

limit = 40




# North America
# US genre
artist_name = []
artist_id = []
genres_US = {}

artist_name, artist_id = search_artist(limit, "US",  artist_name, artist_id)
genres_US = search_genre(artist_id, genres_US)
#print ("us ")


# Canada genre
artist_name = []
artist_id = []
genres_CA = {}

artist_name, artist_id = search_artist(limit, "CA",  artist_name, artist_id)
genres_CA = search_genre(artist_id, genres_CA)

# Mexico genre
artist_name = []
artist_id = []
genres_MX = {}

artist_name, artist_id = search_artist(limit, "MX",  artist_name, artist_id)
genres_MX = search_genre(artist_id, genres_MX)

total_genre = {}
total_genre = genre_total(genres_US, total_genre)
total_genre = genre_total(genres_MX, total_genre)
total_genre = genre_total(genres_CA, total_genre)


total_genre = dict(sorted(total_genre.items(), key=lambda x:x[1], reverse=True))
total_genre = dict(list(total_genre.items())[:5])

print ("Genre in three largest countries of North America",total_genre)


# INDIA
artist_name = []
artist_id = []
genres_IN = {}

artist_name, artist_id = search_artist(limit, "IN",  artist_name, artist_id)
genres_IN = search_genre(artist_id, genres_IN)


# Indonesia
artist_name = []
artist_id = []
genres_ID = {}

artist_name, artist_id = search_artist(limit, "ID",  artist_name, artist_id)
genres_ID = search_genre(artist_id, genres_ID)

# Pakistan
artist_name = []
artist_id = []
genres_PK = {}

artist_name, artist_id = search_artist(limit, "PK",  artist_name, artist_id)
genres_PK = search_genre(artist_id, genres_PK)

total_genre_Asia = {}
total_genre_Asia = genre_total(genres_IN, total_genre_Asia)
total_genre_Asia = genre_total(genres_ID, total_genre_Asia)
total_genre_Asia = genre_total(genres_PK, total_genre_Asia)


total_genre_Asia = dict(sorted(total_genre_Asia.items(), key=lambda x:x[1], reverse=True))
total_genre_Asia = dict(list(total_genre_Asia.items())[:5])

print ("Genre in three largest countries of Asia",total_genre_Asia)


# Europe
# Germany
artist_name = []
artist_id = []
genres_DE = {}

artist_name, artist_id = search_artist(limit, "DE",  artist_name, artist_id)
genres_DE = search_genre(artist_id, genres_DE)


# United Kingdom
artist_name = []
artist_id = []
genres_GB = {}

artist_name, artist_id = search_artist(limit, "GB",  artist_name, artist_id)
genres_GB = search_genre(artist_id, genres_GB)

# France
artist_name = []
artist_id = []
genres_FR = {}

artist_name, artist_id = search_artist(limit, "FR",  artist_name, artist_id)
genres_FR = search_genre(artist_id, genres_FR)


total_genre_Europe = {}
total_genre_Europe = genre_total(genres_DE, total_genre_Europe)
total_genre_Europe = genre_total(genres_GB, total_genre_Europe)
total_genre_Europe = genre_total(genres_FR, total_genre_Europe)


total_genre_Europe = dict(sorted(total_genre_Europe.items(), key=lambda x:x[1], reverse=True))
total_genre_Europe = dict(list(total_genre_Europe.items())[:5])

print ("Genre in three largest countries of Europe", total_genre_Europe)


# South America
# Brazil
artist_name = []
artist_id = []
genres_BR = {}

artist_name, artist_id = search_artist(limit, "BR",  artist_name, artist_id)
genres_BR = search_genre(artist_id, genres_BR)


# Columbia
artist_name = []
artist_id = []
genres_CO = {}

artist_name, artist_id = search_artist(limit, "CO",  artist_name, artist_id)
genres_CO = search_genre(artist_id, genres_CO)

# Argentina
artist_name = []
artist_id = []
genres_AR = {}

artist_name, artist_id = search_artist(limit, "AR",  artist_name, artist_id)
genres_AR = search_genre(artist_id, genres_AR)


total_genre_SA = {}
total_genre_SA = genre_total(genres_BR, total_genre_SA)
total_genre_SA = genre_total(genres_CO, total_genre_SA)
total_genre_SA = genre_total(genres_AR, total_genre_SA)


total_genre_SA = dict(sorted(total_genre_SA.items(), key=lambda x:x[1], reverse=True))
total_genre_SA = dict(list(total_genre_SA.items())[:5])

print ("Genre in three largest countries of South America", total_genre_SA)



# Africa
# Nigeria
artist_name = []
artist_id = []
genres_NG = {}

artist_name, artist_id = search_artist(limit, "NG",  artist_name, artist_id)
genres_NG = search_genre(artist_id, genres_NG)


# Ethiopia
artist_name = []
artist_id = []
genres_ET = {}

artist_name, artist_id = search_artist(limit, "ET",  artist_name, artist_id)
genres_ET = search_genre(artist_id, genres_ET)

# Egypt 
artist_name = []
artist_id = []
genres_EG = {}

artist_name, artist_id = search_artist(limit, "EG",  artist_name, artist_id)
genres_EG = search_genre(artist_id, genres_EG)


total_genre_AF = {}
total_genre_AF = genre_total(genres_NG, total_genre_AF)
total_genre_AF = genre_total(genres_ET, total_genre_AF)
total_genre_AF = genre_total(genres_EG, total_genre_AF)


total_genre_AF = dict(sorted(total_genre_AF.items(), key=lambda x:x[1], reverse=True))
total_genre_AF = dict(list(total_genre_AF.items())[:5])

print ("Genre in three largest countries of Africa",total_genre_AF)

genre_list = [total_genre, total_genre_Asia, total_genre_Europe, total_genre_SA, total_genre_AF]


df = pd.DataFrame.from_dict(genre_list )

df = df.rename(index={'0': 'North America', '1': 'Asia', '2': 'Europe', '3':'South America', '4':'Africa'})
  
df = df.fillna(0)

df.to_csv("Genres .csv", index_label="genre")
```
